### Generating a regular price distribution infographic

Step 1: Convert regular price into numeric

Pre-step: Dealing with missing prices

```python
nullrows = df.isnull().any(axis=1)
df.loc[nullrows,:]
```

```python
df.dropna(axis=0, how='any', inplace = True) #928 observations
#df.loc[nullrows,:] (no more NA observations)
```

```python
df
```

```python
disc_prices = df.disc_prices.tolist()
reg_prices = df.reg_prices.tolist()
```

```python
disc_prices = [item.replace('$','') for item in disc_prices]
reg_prices = [item.replace('$','') for item in reg_prices]
disc_prices
```

```python
reg_prices
```

```python
disc_prices = [round(float(item), 2) for item in disc_prices]
reg_prices = [round(float(item), 2) for item in reg_prices]
disc_prices
```

```python
authors = df.authors.tolist()
authors
```

```python
authors = [item.replace('by\xa0','') for item in authors]
authors
```

```python
cats = df.categories.to_list()
```

```python
cats = [item.replace('nbsp','') for item in cats]
cats = [item.replace('Used&','') for item in cats]
cats
```

```python
titles = df.titles.tolist()
titles
```

```python
clean_df = pd.DataFrame(
    {'author': authors,
     'reg_price': reg_prices,
     'disc_price': disc_prices,
     'categories': cats,
     'title': titles
    })
```

```python
clean_df
```

```python
import seaborn as sns
```

```python
sns.distplot(clean_df['reg_price'], hist = False, kde = True,
                 kde_kws = {'shade': True,'linewidth': 3},
                 label = categories)

plt.title('Density Plot with Regular Price of Used Books')
plt.xlabel('Price (in $)')
plt.ylabel('Density')
plt.savefig('regprice.png',bbox_inches='tight')
```

```python
sns.distplot(clean_df['disc_price'], hist = False, kde = True,
                 kde_kws = {'shade':True,'linewidth': 3},
                 label = categories)

plt.title('Density Plot with Discounted Price of Used Books')
plt.xlabel('Price (in $)')
plt.ylabel('Density')
plt.savefig('discprice.png',bbox_inches='tight')
```

```python
difference = []

zip_object = zip(reg_prices, disc_prices)
for list1_i, list2_i in zip_object:
    difference.append(list1_i-list2_i)

difference = [round(float(d), 2) for d in difference]
difference
```

```python
percentage = []

zip_object = zip(difference,reg_prices)
for list1_i, list2_i in zip_object:
    percentage.append(abs((list1_i/list2_i)*100))

percentage = [round(float(p), 2) for p in percentage]
percentage
```

```python
disc_df = pd.DataFrame(
    {'discount_amount': difference,
     'discount_percentage': percentage
    })
disc_df
```

```python
sns.distplot(disc_df['discount_percentage'], hist = False, kde = True,
                 kde_kws = {'shade':True,'linewidth': 3})

plt.title('Density Plot of Discount Percentage of Used Books')
plt.xlabel('Discount Percentage (in %)')
plt.ylabel('Density')
plt.savefig('discperc.png',bbox_inches='tight')
```
